```python
# Activity 5

# Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read Data
housing_df = pd.read_csv("C:\\Users\\Gabe\\Documents\\Bellevue University\\Data Preparation\\Week 3 & 4\\Boston_housing.csv")
```

```python
# 10 records & total records
print("First 10 Records of DataFrame")
print(housing_df.head(10))
print("")
print("Number of Records")
print(housing_df.shape)
print("")
```

```python
# DataFrame exlcuding CHAS, NOX, B, LSTAT
housing_df_smaller = housing_df[['CRIM','ZN','INDUS','RM','AGE','DIS','RAD','TAX','PTRATIO','PRICE']]
```

```python
# Last 7 records of new DataFrame
print("Last Seven Records of New DataFrame")
print(housing_df_smaller.tail(7))
print("")
```

```python
# Histograms of all columns of new DataFrame
print("Histograms of all columns in New DataFrame")
for column in housing_df_smaller.columns:
    plt.title(column+ " of Boston")
    plt.hist(housing_df_smaller[column], bins=20)
    plt.show()
```

```python
# Scatter plot of crime rate against price
print("")
plt.scatter(housing_df_smaller['CRIM'], housing_df_smaller['PRICE'])
plt.title("Crime Rate VS Price")
plt.xlabel("Crime")
plt.ylabel("Price")
plt.show()
```

```python
# Plot using log10(crime) against price
plt.scatter(np.log10(housing_df_smaller['CRIM']),housing_df_smaller['PRICE'])
plt.title("Log of Crime Rate vs. Price")
plt.xlabel("Log of Crime rate")
plt.ylabel("Price")
plt.show()
```

```python
# Mean Rooms per Dwelling
print("Mean Rooms per Dwelling")
print(housing_df_smaller['RM'].mean())
print("")

# Median Age
print("Median Age")
print(housing_df_smaller['AGE'].median())
print("")

# Mean Distance to Five Boston Employment Centers
print("Mean Distance to Five Boston Employment Centers")
print(housing_df_smaller['DIS'].mean())
print("")

# Percentage of Houses with a low price (<$20,000)
print("Percentage of Houses with a lowe price (<$20,000)")
# creating an array of booleans containing if the price is less than 20
houses_lower_price = housing_df_smaller['PRICE'] < 20
# Mean of All True (1's) made into percentage
houses_percent_low = houses_lower_price.mean() * 100
print(houses_percent_low, "%")
```
