```python
import numpy as np
import pandas as pd
df = pd.read_csv("C:\\Users\\Gabe\\Documents\\Bellevue University\\Data Preparation\\Week 7 & 8\\candyhierarchy2017.csv",encoding="ISO-8859-1")
df.head()
```

```python
df['Q3: AGE'] = df['Q3: AGE'].astype(str)
```

```python
# Chapter 7 - Filter out missing data
df = df[~df['Q3: AGE'].str.contains("1/2")]
df = df[~df['Q3: AGE'].str.contains("312")]
df = df[~df['Q3: AGE'].str.contains("1000")]
df = df[~df['Q3: AGE'].str.contains("24-5")]
df = df[~df['Q3: AGE'].str.contains("45-55")]
df = df[~df['Q3: AGE'].str.contains('Old enough')]
df = df[~df['Q3: AGE'].str.contains('46 Halloweens.')]
df = df[~df['Q3: AGE'].str.contains('59 on the day after Halloween')]
df = df[~df['Q3: AGE'].str.contains('5u')]
df = df[~df['Q3: AGE'].str.contains('60+')]
df = df[~df['Q3: AGE'].str.contains('ancient')]
df = df[~df['Q3: AGE'].str.contains('Enough')]
df = df[~df['Q3: AGE'].str.contains('hahahahaha')]
df = df[~df['Q3: AGE'].str.contains('I can remember when Java was a cool new language')]
df = df[~df['Q3: AGE'].str.contains('Many')]
df = df[~df['Q3: AGE'].str.contains('MY NAME JEFF')]
df = df[~df['Q3: AGE'].str.contains('no')]
df = df[~df['Q3: AGE'].str.contains('OLD')]
df = df[~df['Q3: AGE'].str.contains('old')]
df = df[~df['Q3: AGE'].str.contains('older than dirt')]
df = df[~df['Q3: AGE'].str.contains('Over 50')]
df = df[~df['Q3: AGE'].str.contains('See question 2')]
df = df[~df['Q3: AGE'].str.contains('sixty-nine')]
df = df[~df['Q3: AGE'].str.contains('your mom')]
df = df[~df['Q3: AGE'].str.contains('?', regex = False)]
df.drop([0], inplace=True)
df = df[~df['Q3: AGE'].str.contains('nan')]
df = df.reset_index(drop=True)
```

```python
# Chapter 7 - Replace Values
df['Q12: MEDIA [Daily Dish]'] = df['Q12: MEDIA [Daily Dish]'].fillna(0)
df['Q12: MEDIA [Science]'] = df['Q12: MEDIA [Science]'].fillna(0)    
df['Q12: MEDIA [ESPN]'] = df['Q12: MEDIA [ESPN]'].fillna(0) 
df['Q12: MEDIA [Yahoo]'] = df['Q12: MEDIA [Yahoo]'].fillna(0)
df.head()
```

```python
# Chapter 7 - Replace Values

for column in df:
    df[column] = df[column].astype(str)
    df[column] = df[column].str.replace('MEH', '2')
    df[column] = df[column].str.replace('DESPAIR', '1')
    df[column] = df[column].str.replace('JOY', '3')
    #df[column] = df[column].str.replace('nan', '0')
    
df['Q1: GOING OUT?'] = df['Q1: GOING OUT?'].str.replace('0','No')
```

```python
df['Q3: AGE']=df['Q3: AGE'].astype('float64')
```
