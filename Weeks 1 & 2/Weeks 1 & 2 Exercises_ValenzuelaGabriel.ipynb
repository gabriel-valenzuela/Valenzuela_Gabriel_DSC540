```python
# 3 - Data Warngling Activites
import math
from itertools import permutations, dropwhile, zip_longest

# Activity 1

LIMIT = 100
random_number_list = [np.random.randint(0, LIMIT) for x in range(0, LIMIT)]
print(random_number_list)

divisible_list = [num for num in random_number_list if num % 3 == 0]
print(divisible_list)

diff = len(random_number_list) - len(divisible_list)
print(diff)

diff_list = []
for i in range(0,100):
    random_number_list = [np.random.randint(0, LIMIT) for x in range(0, LIMIT)]
    divisible_list = [num for num in random_number_list if num % 3 == 0]
    diff = len(random_number_list) - len(divisible_list)
    diff_list.append(diff)
print(diff_list)

print(np.mean(diff_list))

# Activity 2

with open ('C:\\Users\\Gabe\\PycharmProjects\\DataPreparation_Week1&2\\PP.txt', 'r') as file:
    chapter = file.read()

print(len(chapter))

chapter = chapter.replace('\n', '')
cleaned = ""
for char in chapter:
    if char == " ":
        cleaned += char
    elif char.isalnum():
        cleaned += char
    else:
        cleaned += " "

word_list = cleaned.split()
print(len(word_list))

unique_words = dict.fromkeys(word_list)
print(len(list(unique_words.keys())))

for word in word_list:
    if unique_words[word] is None:
        unique_words[word] = 1
    else:
        unique_words[word] += 1

top_25_words = sorted(unique_words.items(),key=lambda key_val_tuple: key_val_tuple[1], reverse=True)
print(top_25_words[:25])

# Activity 3

permutations(range(3))
for num in permutations(range(3)):
    print(num)
    assert isinstance(num, tuple)
for num in permutations(range(3)):
    print(list(dropwhile(lambda x: x <= 0, num)))

def create_num(num_stack):
    created_num = 0
    for i in range(0, len(num_stack)):
        created_num += (num_stack.pop() * math.pow(10,i))
    return created_num


for num in permutations(range(3)):
    num_stack = list(dropwhile(lambda x: x <= 0, num))
    print(create_num(num_stack))

# Activity 4


def return_dict_from_csv_line(header, line):
    zipped_line = zip_longest(header,line,fillvalue=None)
    ret_dict = {kv[0]: kv[1] for kv in zipped_line}
    return ret_dict


fd = open("C:\\Users\\Gabe\\PycharmProjects\\DataPreparation_Week1&2\\sales_record.csv","r")
first_line = fd.readline()
header = first_line.replace("\n","").split(",")
for i, line in enumerate(fd):
    line = line.replace("\n","").split(",")
    d = return_dict_from_csv_line(header, line)
    print(d)
    if i > 10:
        break
```
